# Crear Data Frames

```python
df = pd.DataFrame({
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10)
})
df
```

```python
df.mean()
```

```python
df.sum()
```

```python
df.mean(axis='columns')
```

```python
df2 = pd.DataFrame({
    'key': ['A','B','C','A','B','C'],
    'values': range(6)
}, columns = ['key', 'values'])
```

```python
df2
```

```python
df2.groupby('key')
```

```python
df2.groupby('key').mean()
```
